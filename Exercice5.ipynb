```python
import pandas as pd

file_path = '/content/drive/MyDrive/dm_office_sales.csv'
df = pd.read_csv(file_path)
display(df.head())
```

Cette cellule charge le fichier CSV `dm_office_sales.csv` dans un DataFrame pandas nommé `df` et affiche les cinq premières lignes pour un aperçu des données.

```python
import seaborn as sns
sns.lmplot(data=df, x="salary", y="sales",hue="level of education")
```

Ce graphique de dispersion (`lmplot`) de Seaborn visualise la relation entre le salaire (`salary`) et les ventes (`sales`), avec différentes couleurs représentant le niveau d'éducation (`level of education`).

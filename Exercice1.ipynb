```python
matrice5by5=np.arange(1,26,1).reshape(5,5)
print(matrice5by5)
```

```python
ecarttype=np.std(matrice5by5)
print(ecarttype)
```

Calcule l'écart-type des valeurs de la matrice, ce qui mesure la dispersion des données.

```python
column_sum=np.sum(matrice5by5,axis=0)
print(np.array(column_sum))
```

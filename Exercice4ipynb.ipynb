```python
periode = ['1 Mois', '3 Mois', '6 Mois', '1 An', '2 An', '3 An', '5 An', '7 An', '10 An', '20 An', '30 An']
juillet16_2007 = [4.75, 4.98, 5.08, 5.01, 4.89, 4.89, 4.95, 4.99, 5.05, 5.21, 5.14]
juillet16_2020 = [0.12, 0.11, 0.13, 0.14, 0.16, 0.17, 0.28, 0.46, 0.62, 1.09, 1.31]
```

```python
import matplotlib.pyplot as plt

fig , ax1 =plt.subplots()
ax1.plot(periode,juillet16_2007)
color = 'tab:blue'
ax1.set_xlabel('periode')
ax1.set_ylabel('j-2007')
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:green'
ax2.plot(periode,juillet16_2020,color=color)
ax2.set_ylabel('j-2020')
ax2.tick_params(axis='y', labelcolor=color)
ax1.grid(True)




```

```python
#Utiliser des subplots pour créer la figure qui montre la courbe de rendement de chaque
#année
```

```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))


ax1.plot(periode, juillet16_2007, marker='o', color='tab:blue')
ax1.set_xlabel('Periode')
ax1.set_ylabel('Rendement')
ax1.set_title('Courbe de rendement (16 Juillet 2007)')
ax1.tick_params(axis='x', rotation=45)
ax1.grid(True)


ax2.plot(periode, juillet16_2020, marker='o', color='tab:green')
ax2.set_xlabel('Periode')
ax2.set_ylabel('Rendement')
ax2.set_title('Courbe de rendement (16 Juillet 2020)')
ax2.tick_params(axis='x', rotation=45)
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python


fig, ax1 = plt.subplots()

color1 = 'tab:blue'
ax1.set_xlabel('Période')
ax1.set_ylabel('Rendement 2007', color=color1)
ax1.plot(periode, juillet16_2007, color=color1, marker='o', label='16 Juillet 2007')
ax1.tick_params(axis='y', labelcolor=color1)
ax1.grid(True)

ax2 = ax1.twinx()
color2 = 'tab:green'
ax2.set_ylabel('Rendement 2020', color=color2)
ax2.plot(periode, juillet16_2020, color=color2, marker='s', label='16 Juillet 2020')
ax2.tick_params(axis='y', labelcolor=color2)

plt.title('Comparaison des courbes de rendement (Double axe Y)')
fig.tight_layout()
plt.show()
```

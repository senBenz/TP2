# Q8- voici la colonne montant qui est calculee sous la formule suivante


```
montant = prix_unitaire × quantite × (1 − remise_pct/100).
```



````python
import math
stats=0


def resume_statistiques(df, colonne):
    stats={
        'n ': len(df[colonne]),
        ' somme  ':float(df[colonne].sum()),
        'moyenn ':float(df[colonne].mean()),
        'median ':df[colonne].median(),
        'ecart_type ':math.sqrt(df[colonne].var()),
        'Q1 ':float(df[colonne].quantile(0.25)),
        'Q3 ':float(df[colonne].quantile(0.75)),
        'IQR ':float(df[colonne].quantile(0.75)-df[colonne].quantile(0.25))
    }
    return stats








print(resume_statistiques(df,'montant'))
````

```python
import pandas as pd
import numpy as np

np.random.seed(101)
mydata = np.random.randint(0,101,(4,3))
myindex = ['CANADA','USA','FRANCE','UK']
mycolumns = ['Jan','Feb','Mar']
dframe=pd.DataFrame(mydata,index=myindex,columns=mycolumns)
print(dframe)
```

### Explication
Cette cellule initialise un DataFrame pandas `dframe` avec des valeurs entières aléatoires comprises entre 0 et 100. Les lignes représentent des pays et les colonnes les mois de janvier à mars.

```python
print(dframe.describe())
```

### Explication
La méthode `.describe()` génère des statistiques descriptives (moyenne, écart-type, min, max, quartiles) pour chaque colonne numérique du DataFrame.

```python
dframe.drop('FRANCE')
```

### Explication
L'opération `.drop('FRANCE')` renvoie une copie du DataFrame sans la ligne 'FRANCE'. Notez que sans l'argument `inplace=True`, le DataFrame original `dframe` reste inchangé.

```python
dframe.loc['USA']
```

### Explication
L'attribut `.loc['USA']` est utilisé pour sélectionner et afficher toutes les données relatives à la ligne 'USA' sous forme de Série.

```python
dframe.loc[:,'Jan']
```

### Explication
Ici, `.loc[:,'Jan']` permet de sélectionner l'intégralité de la colonne 'Jan' pour tous les pays présents dans l'index.

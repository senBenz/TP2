```python
import pandas as pd

file_path = '/content/drive/MyDrive/movie_scores.csv'
df = pd.read_csv(file_path)
display(df.head())
```

On charge le fichier CSV dans un DataFrame pandas et on affiche les 5 premières lignes pour inspecter les données.

```python
df.isnull()
```

La méthode `isnull()` retourne un booléen pour chaque cellule : `True` si la donnée est manquante (NaN), `False` sinon.

```python
df['first_name'].dropna()
```

Ici, on extrait la colonne 'first_name' en supprimant les valeurs nulles, ce qui permet de ne garder que les prénoms valides.

```python
df['post_movie_score'].sort_values(ascending=False).dropna()
```

Cette commande trie les scores par ordre décroissant et élimine les valeurs manquantes pour identifier les meilleures performances.

```python
df.fillna('Nouvelle_valeur')
```

`fillna()` est utilisé pour remplacer toutes les valeurs manquantes par un texte par défaut, évitant ainsi les calculs sur des données vides.

```python
df.dropna(axis='columns')




```

`dropna(axis='columns')` supprime toute colonne contenant au moins une valeur nulle. Comme une ligne est entièrement vide dans ce dataset, toutes les colonnes ont été supprimées.
